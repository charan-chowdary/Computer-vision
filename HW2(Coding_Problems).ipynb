```python

#Problem 3

import matplotlib.pyplot as plt
import pandas as pd

# Function to perform linear regression using gradient descent
def linear_regression_gradient_descent(learning_rate, epochs, X, Y):
    # Initialize parameters
    m = 0  # Slope
    c = 0  # Intercept
    eta = learning_rate
    n = len(X)

    for _ in range(epochs):
        # Predicted Y values
        Y_pred = m * X + c

        # Compute gradients
        d_m = (-2/n) * sum(X * (Y - Y_pred))
        d_c = (-2/n) * sum(Y - Y_pred)

        # Update parameters using gradients
        m -= eta * d_m
        c -= eta * d_c

    return m * X + c

# Start of the program
# Reading data from a CSV file
data = pd.read_csv("https://raw.githubusercontent.com/charan-chowdary/HW2/main/HW2_linear_data-1.csv", header=None)

# Extracting values for X and Y from the dataset
X_values = data.iloc[:, 0]
Y_values = data.iloc[:, 1]
plt.scatter(X_values, Y_values, s=15)

# Performing linear regression with gradient descent
predicted_Y = linear_regression_gradient_descent(learning_rate=0.0001, epochs=1000, X=X_values, Y=Y_values)

# Plotting the linear regression line
plt.plot(X_values, predicted_Y, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()
```

```python
#Problem 4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def cubic_regression_gradient_descent(X, Y, learning_rate=1e-6, epochs=10000):
    # Initialize coefficients for the cubic function
    a = 0.0
    b = 0.0
    c = 0.0
    d = 0.0
    data_len = len(X)

    # Perform Gradient Descent to optimize a, b, c, and d
    for epoch in range(epochs):
        # Calculate predictions using the cubic function
        Y_pred = a * (X ** 3) + b * (X ** 2) + c * X + d

        # Calculate Mean Squared Error (MSE)
        mse = np.mean((Y - Y_pred) ** 2)

        # calculating gradients
        a_grad = (-2 / data_len) * np.sum((X ** 3) * (Y - Y_pred))
        b_grad = (-2 / data_len) * np.sum((X ** 2) * (Y - Y_pred))
        c_grad = (-2 / data_len) * np.sum(X * (Y - Y_pred))
        d_grad = (-2 / data_len) * np.sum(Y - Y_pred)

        # Updating paramaters using learning rate and gradient
        a -= learning_rate * a_grad
        b -= learning_rate * b_grad
        c -= learning_rate * c_grad
        d -= learning_rate * d_grad

    return a, b, c, d

# Load the data
data = pd.read_csv("https://raw.githubusercontent.com/charan-chowdary/HW2/main/HW2_nonlinear_data.csv")
X = data['X'].values
Y = data['Y'].values

# Perform cubic regression with gradient descent
a, b, c, d = cubic_regression_gradient_descent(X, Y)

# Plotting the results
X_plot = np.linspace(min(X), max(X), 100)
Y_plot = a * (X_plot ** 3) + b * (X_plot ** 2) + c * X_plot + d
plt.scatter(X, Y)
plt.scatter(X_plot, Y_plot, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Non-Linear Regression")
plt.legend()
plt.show()

# Output final coefficients and intercept
print(f"Final coefficients:")
print(f"a : {a}")
print(f"b : {b}")
print(f"c : {c}")
print(f"d : {d}")
```
